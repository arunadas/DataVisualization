```python
import matplotlib.pyplot as plt
from matplotlib import style
import random 

style.use('custom.mplstyle')

for label in range(12):
    x = []
    y = []
    
    for i in range(1,10):
        ys = random.randrange(0,15)
        xs = i
        x.append(xs)
        y.append(ys)
        
    plt.plot(x, y, label = label)

plt.legend()
plt.show()
#print(style.available)
```
