# Label, title buttons, legends

```python
x=[1,3,4,5,7]
y=[3,4,2,5,6]
y2=[5,7,4,3,1]

plt.plot(x,y, label='Initial Line')
plt.plot(x,y2, label='New Line')

plt.xlabel('Plot Number')
plt.ylabel('Random #')

plt.title('Ploting x/y points')
plt.legend()

plt.show()
```
